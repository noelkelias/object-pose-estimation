# Data export

Export RGB-D frames + ground truth. Picks the best available backend: **Isaac Sim** → **Gazebo** → **`data/synthetic_exports/`**.

Run after `setup.ipynb`. Colab does not run simulators — export locally, then upload exports to Drive.

```python
import os
import runpy

_cwd = os.getcwd()
_utils = os.path.join(_cwd, "colab_utils.py")
if not os.path.isfile(_utils):
    _utils = os.path.join(_cwd, "notebooks", "colab_utils.py")
runpy.run_path(_utils)

from notebooks.colab_utils import display_path, setup_notebook

PROJECT_ROOT = setup_notebook()
print("Project root:", display_path(PROJECT_ROOT, project_root=PROJECT_ROOT))
```

## Pick export backend

Same priority as `load_frame`: Isaac if installed, else Gazebo, else synthetic demo frames.

```python
from data.frame import describe_frame_sources, export_search_dirs
from data.gazebo import gazebo_available
from data.isaac import isaac_available

for line in describe_frame_sources(PROJECT_ROOT):
    print(line)

isaac_dir, gazebo_dir, synthetic_dir = export_search_dirs(PROJECT_ROOT)
isaac_ok, isaac_msg = isaac_available()
gz_ok, gz_msg = gazebo_available()

if isaac_ok:
    BACKEND, backend_msg, out_dir = "isaac", isaac_msg, isaac_dir
elif gz_ok:
    BACKEND, backend_msg, out_dir = "gazebo", gz_msg, gazebo_dir
else:
    BACKEND = "synthetic"
    backend_msg = "Isaac Sim and Gazebo not available; writing bundled demo frames"
    out_dir = synthetic_dir

print()
print(f"Selected backend: {BACKEND}")
print(f"Reason: {backend_msg}")
print(f"Output dir: {display_path(out_dir, PROJECT_ROOT)}")
```

## Export frames

```python
NUM_FRAMES = 20
FORCE = False

from data.frame import export_sim_frames

_ = export_sim_frames(BACKEND, out_dir, num_frames=NUM_FRAMES, force=FORCE)
```
